```python
import matplotlib.pyplot as plt
```

```python
# 라인 그래프 출력
plt.plot([1,2,3,4])
plt.show()
```

```python
# 라인 그래프 x축의 값 y 축의 값 지정
plt.plot([1,2,3,4], [1,200,2,8])
plt.show()
```

```python
x = [1,2,3,4]
y = [10,3,4,9]

plt.plot(x, y, "--r")
plt.plot(x, [4,2,6,4], "--y")
plt.xlabel("Class") #x축의 이름
plt.ylabel("Count") #y축의 이름
plt.title("Class & Count") #그래프의 제목
plt.legend(["A", "B"], loc=0) # 범례 추가
plt.show()
```

```python
# subplot
plt.subplot(1 , 2, 1)
plt.plot([1,2,3,4])
plt.subplot(1, 2, 2)
plt.plot([8,7,6,5])
plt.show()
```

1. 임의의 그래프 4개를 2X2 subplot 출력

```python
plt.subplot(2, 2, 1)
plt.plot([3,5,6,7])
plt.legend(["A"])
plt.subplot(2,2,2)
plt.plot([1,2,3,4])
plt.subplot(2,2,3)
plt.plot([9,4,6,3])
plt.subplot(2,2,4)
plt.bar([1,2,3,4],[10,2,6,1])
plt.show()
```

```python
# 막대 그래프 
x = [1,2,3,4]
y = [80,40,70,30]

plt.bar(x, y)
plt.show()
```

```python
plt.bar(x, y, color=["r", "y", "b"], width=0.8)
plt.xticks(x, ["maths", "seience", 'history', 'english'])
plt.show()
```

```python
import numpy as np
```

```python
# 산점도 그래프 
x = np.random.rand(50)
y = np.random.rand(50)
print(x)
```

```python
plt.scatter(x, y)
plt.show()
```

```python
# 산점도 그래프 옵션 추가 (영역 확장, 색상 변경, 투명도 조절)

color = np.random.rand(50)
area = (20 * np.random.rand(50))**2

plt.scatter(x, y, s=area, c=color, alpha = 0.7)
plt.show()
```
