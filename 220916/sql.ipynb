```python
import pymysql
import pandas as pd
```

```python
## 어느 데이터베이스에 접속을 할지 설정
_db = pymysql.connect(
    user = 'root', 
    password = "1234", 
    host = "localhost", 
    db = "ubion", 
    port = 3306
)
```

```python
## cursor는 sql구문을 실행 시키기 위해 만들어야되는 변수
## DictCursor -> sql에서 불러온 데이터를 딕셔너리 형태로 받아온다.
cursor = _db.cursor(pymysql.cursors.DictCursor)
```

```python
## select 문 사용
sql = """select * from EMP 
        where DEPTNO = 30"""

cursor.execute(sql)
result = cursor.fetchall()
print(result)
```

```python
## 데이터베이스에서 받아온 리스트 안에 딕셔너리 형태의 데이터를 
## 데이터프레임으로 변형
pd.DataFrame(result)
```

```python
## sql CRUD
## C : Create -> insert
## R : Read -> select -> 결과물 반환 -> fetchall()
## U : Update -> update
## D : Delete -> delete
## 데이터베이스 변환 -> execute() -> commit()
sql = """
        insert into DEPT values 
        (100, "A", "seoul")
"""

cursor.execute(sql)
```

```python
_db.commit()
```

```python
sql = """
        insert into DEPT values (%s, %s, %s)
"""
deptno = input("DEPTNO = ")
dname = input("DNAME = ")
loc = input("LOC = ")
values = [deptno, dname, loc]

cursor.execute(sql, values)
_db.commit()
```

```python
sql = """
        select 
        `Units Sold`, 
        `Unit Price`, 
        `Unit Cost`, 
        `Total Revenue`, 
        (`Units Sold` * `Unit Price`) as 총수익, 
        `Total Cost`, 
        (`Units Sold` * `Unit Cost`) as 총비용
        from 
        `sales records`
"""
test.executeall(sql)
```

```python
## 반올림
sql = """
        select 
        `Total Cost`, 
        round(`Total Cost`, 1)
        from 
        `sales records`
"""
test.executeall(sql)
```

```python
## 그룹화 
## Country, Item Type 그룹화
## Units Sold 평균, 합계 파생변수(AVG, SUM) 2개 생성
## Country 오름차순 정렬
## 컬럼 Country, Item Type, AVG, SUM
sql = """
        select 
        `Country`, 
        `Item Type`, 
        AVG(`Units Sold`) as AVG, 
        SUM(`Units Sold`) as SUM 
        from 
        `sales records` 
        group by 
        `Country`, 
        `Item Type` 
        order by 
        `Country` 
        ASC
"""
test.executeall(sql)
```
